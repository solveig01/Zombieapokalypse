```python
from pylab import *

#Tid og steg
N = 100000
t0 = 0
t_slutt = 100
dt = (t_slutt - t0)/N

#Konstanter
f = 0.01        #Naturlig fødselsrate
d = 0.02        #Naturlig dødsrate 
inn = 0.04      #Innvandringsrate


#Matriser
M = zeros(N)
Mder = zeros(N)
t = zeros(N)

#Initialbetingelser
M[0] = 500      #Mennesker i Alexandria ved start
t[0] = 0

#Eulers metode
for i in range(N-1):
    Mder[i] = f*M[i] + inn*M[i] - d*M[i]
    M[i+1] = M[i] + Mder[i]*dt
    t[i+1] = t[i] + dt

#Plotting
plot(t,M)
axhline(y=0, color='black')
axvline(x=0, color='black')
grid()
```

```python
from pylab import *

#Tid og steg
N = 100000
t0 = 0
t_slutt = 1000
dt = (t_slutt - t0)/N

#Konstanter mennesker
f = 0.01              #Naturlig fødselsrate
inn = 0.04            #Innvandringsrate
d_alder = 0.002       #Naturlig dødsrate 
d_syk = 0.03          #Dødsrate grunnet sykdom
d_skade = 0.01        #Dødsrate grunnet skader
d_zombie = 0.02       #Dødsrate for drepte av zombie
b = 1000              #Bæreevne

#Konstanter zombier
s = 0.0002            #Zombienes spiserate
drept = 0.5           #Dødsrate for zombier drept av mennesker

#Matriser
M = zeros(N)
Mder = zeros(N)
Z = zeros(N)
Zder = zeros(N)
t = zeros(N)

#Initialbetingelser
M[0] = 500
Z[0] = 500
t[0] = 0

#Eulers metode
for i in range(N-1):
    Mder[i] = M[i] * ((f + inn - d_zombie)*Z[i]) * (1-(M[i]/b))
    Zder[i] = Z[i] * (s*M[i]) * (drept*M[i]*Z[i])
    M[i+1] = M[i] + Mder[i]*dt
    Z[i+1] = Z[i] + Zder[i]*dt
    t[i+1] = t[i] + dt
    
#Plotting
plot(t,M)
plot(t,Z)
legend(['Mennesker','Zombier'])
axhline(y=0, color='black')
axvline(x=0, color='black')
grid()
```
